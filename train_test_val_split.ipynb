```python
import pandas as pd 
import random 
import seaborn as sns
```

```python
ground=pd.read_csv('input/Ground Truth.csv', index_col=0)
```

```python
random.seed(42)
random_assignment = random.choices(population=['train', 'test','val'],
                                   weights=[0.8,0.1,0.1], 
                                   k=len(ground))
```

```python
ground['split'] = random_assignment
```

```python
sns.countplot(ground, x='split', hue='y')
```

```python
ground.to_csv('processed/ground_split.csv', index=0)
```
